```python
# Optimal pandas import
import warnings
warnings.filterwarnings("ignore")
import pandas as pd

pd.options.display.max_colwidth = -1
pd.options.display.max_rows = 999
from unidecode import unidecode
```

# Data Electoral 2015

## Nombres de listas

```python
COD_PARTIDOS = pd.read_csv('./../data/codigos/COD_PARTIDOS_2015.csv', ';', encoding = 'latin-1')[[u'Codigo Partido', u'Denominacion']]
COD_PARTIDOS.columns = [unidecode(col) for col in COD_PARTIDOS.columns] # atender  Esta tilde
COD_PARTIDOS['Denominacion'] = COD_PARTIDOS['Denominacion'].str.strip()
COD_PARTIDOS['Denominacion'] = COD_PARTIDOS['Denominacion'].str.replace('ALIANZA CAMBIEMOS', 'CAMBIEMOS')
COD_PARTIDOS['Denominacion'] = COD_PARTIDOS['Denominacion'].str.replace('CAMBIEMOS BUENOS AIRES', 'CAMBIEMOS')
```

## Nombres Geograficos

```python
COD_PROV = pd.read_csv('./../data/codigos/COD_PROVINCIAS.csv', ';', encoding = 'latin-1').rename(columns = {'NOMBRE': 'PROVINCIA'})
COD_DEP = pd.read_csv('./../data/codigos/COD_DEPARTAMENTOS.csv', ';', encoding = 'latin-1').rename(columns = {'NOMBRE': 'DEPARTAMENTO'})

COD_PROV['PROV_REF_ID'] = COD_PROV['CODIGO PROVINCIA'].astype(int)#.astype(str)
COD_DEP['PROV_REF_ID'] = COD_DEP['CODIGO PROVINCIA'].astype(int)
```

## Presi y Vice

```python
#Load data
PR_2015 = pd.concat([pd.read_csv('./../data/escrutinio/2015/FMESPR_0101.csv', ';'),
                            pd.read_csv('./../data/escrutinio/2015/FMESPR_0202.csv', ';'),
                            pd.read_csv('./../data/escrutinio/2015/FMESPR_0313.csv', ';'),
                            pd.read_csv('./../data/escrutinio/2015/FMESPR_1424.csv', ';')])

PR_2015['ELECCION'] = 'PV_2015'
# #fix
PR_2015['CODIGO VOTOS'] = PR_2015['CODIGO VOTOS'].replace(135, 145)
PR_2015.columns = PR_2015.columns.str.strip()
PR_2015['circuito'] = PR_2015['CODIGO CIRCUITO'].astype(str).str.strip().str.zfill(4)
```

# MERGE

```python
PR_2015 = PR_2015.rename(columns = {'VOTOS': 'VOTOS_PR'})
DN_2015 = DN_2015.rename(columns = {'VOTOS': 'VOTOS_DN'})
GO_2015 = GO_2015.rename(columns = {'VOTOS': 'VOTOS_GO'})
MU_2015 = MU_2015.rename(columns = {'VOTOS': 'VOTOS_MU'})

merged = PR_2015.loc[PR_2015['CODIGO PROVINCIA'] == 2].merge(DN_2015.loc[DN_2015['CODIGO PROVINCIA'] == 2],
                                                            ).merge(GO_2015
                                                                   ).merge(MU_2015)
# merged = merged.dropna()
# set(GO_2015.columns) - set(MU_2015.columns)
```

```python
 # mergear
merged = merged.merge(COD_PARTIDOS, left_on = u'CODIGO VOTOS', right_on = u'Codigo Partido')
# mergear parte geografica
merged = merged.merge(COD_PROV).merge(COD_DEP, on = ['CODIGO DEPARTAMENTO', u'PROV_REF_ID', 'CODIGO PROVINCIA'])
```

```python
import seaborn as sns
cm = sns.light_palette("green", as_cmap=True)
```

```python
dpto_ix = merged.groupby('DEPARTAMENTO')['VOTOS_PR'].sum().sort_values(ascending = False).index
```

```python
df = merged.groupby(['DEPARTAMENTO','CODIGO VOTOS', 'Denominacion'])[[col for col in merged.columns if 'VOTOS_' in col]].sum()
```

```python
df_ = df.div(df.mean(1), 0).dropna().round(2)
df_ = df_.loc[df_.index.get_level_values('CODIGO VOTOS').isin([131, 132, 138, 145])]
df_ = df_.reindex(dpto_ix[:50], level = 0)
df_['diff_gob'] = df_['VOTOS_GO'] - df_['VOTOS_PR']
s = df_.sort_values(by = 'diff_gob').style.background_gradient(cmap=cm, axis = 0)
s
```
